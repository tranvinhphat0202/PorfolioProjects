```python
df['gender'].value_counts(normalize = True)
```

```python
plt.figure(figsize = (20, 6))

ax = df["gender"].value_counts().plot(kind = 'bar', color = colors, rot = 0)
ax.set_xticklabels(('Female', 'Male'))

for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x() + 0.25, p.get_height() + 1), ha = 'center', va = 'bottom', color = 'black')
    ax.tick_params(axis = 'both', labelsize = 15)

plt.xlabel('Gender', weight = "bold", color = "#D71313", fontsize = 14)
plt.ylabel('Counts', weight = "bold", color = "#D71313", fontsize = 14)
```

```python
plt.figure(figsize = (20, 6))

counts = df["gender"].value_counts()
explode = (0, 0.1)

counts.plot(kind = 'pie', fontsize = 12, colors = colors, explode = explode, autopct = '%.1f%%')
plt.title('Males Vs Females')
plt.xlabel('Gender', weight = "bold", color = "#2F0F5D", fontsize = 14, labelpad = 20)
plt.ylabel('Counts', weight = "bold", color = "#2F0F5D", fontsize = 14, labelpad = 20)
plt.legend(labels = counts.index, loc = "best")
plt.show()
```

```python
total_sales_per_item = df.groupby('category')['price'].sum()

print(total_sales_per_item)
```

```python
plt.figure(figsize = (16, 6))

total_sales_per_item.sort_values(ascending = False).plot(kind = 'bar', color = sns.color_palette('inferno'), edgecolor = 'black')
plt.xlabel('Items', weight = "bold", color = "#D71313", fontsize = 14, labelpad = 20)
plt.ylabel('\nTotal Sales', weight = "bold", color = "#D71313", fontsize = 14, labelpad = 20);
plt.xticks(rotation = 0, ha = 'center', fontsize = 16)

plt.tight_layout()
plt.show()
```

```python
df_sales = total_sales_per_item.reset_index()

plt.figure(figsize=(10, 6))

sns.set_palette('winter_r')  
explode = [0.2, 0, 0.2, 0.2, 0, 0.2, 0 ,0.2]

plt.pie(df_sales['price'], labels=df_sales['category'], autopct='%1.1f%%', explode = explode)
plt.axis('equal') 
plt.title('Total Sales per Item')

plt.show()
```
